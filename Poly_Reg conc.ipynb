```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
% matplotlib inline
import pandas as pd
```

```python
np.random.seed(42)
```

```python
n_samples=100
```

```python
rng=np.random.randn(n_samples)*100
```

```python
print(rng)
```

```python
X=np.linspace(0,10,100)
```

```python
Y=X**3+rng+100
```

```python
plt.figure(figsize=(10,8))
#plt.plot(X,Y)
plt.scatter(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X.reshape(-1,1),Y)
```

```python
x_pred=lr.predict(X.reshape(-1,1))
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,Y)
plt.plot(X,x_pred)
```

```python
plt.plot(X,x_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_fe = PolynomialFeatures(degree=2)
x_poly=poly_fe.fit_transform(X.reshape(-1,1))
```

```python
lin_reg=LinearRegression()
```

```python
lin_reg.fit(x_poly,Y.reshape(-1,1))
```

```python
y_pred=lin_reg.predict(x_poly)
```

```python
plt.figure(figsize=(10,8));
plt.scatter(X, Y);
plt.plot(X, y_pred);
```
